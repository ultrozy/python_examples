```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (20.0, 20.0)


A = 1000
N = 1000
M = N * A
All = []

for i in range(50):
    R = np.random.randint(0, N, M)
    elm, cnt = np.unique(R, return_counts=True)
    List = [0] * N
    for i in range(len(elm)):
        List[elm[i]] = cnt[i]
    List.sort()
    All.append(List)
All = np.array(All) - A
Average = All.mean(axis = 0)
plt.plot(Average)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import tau

plt.rcParams["figure.figsize"] = (20, 20)
DOTS = 1000
MAXN = 100
```

```python
X = np.linspace(-np.pi/2, np.pi/2, num=DOTS)
Ginv = np.array([[x for x in X] for _ in range(MAXN)])
Ginv[1] = np.pi * np.sin(X) / 2
plt.plot(X, Ginv[0])
plt.plot(X, Ginv[1])

cos2 = np.cos(X)
cos2 *= cos2

kn1 = np.array(Ginv[1])
kn = 0.5 * np.sin(2 * X)

for n in range(2, MAXN):
    Ginv[n] = kn + Ginv[n-2]
    kn1, kn = kn, ((n-1) / n) * (cos2 * kn1)
    
    plt.plot(X, Ginv[n])
plt.show()
```

```python
X = np.linspace(-1, 1, num=DOTS)
Finv = np.array([[x for x in X] for _ in range(MAXN)])
Finv[0] = (2 / np.pi) * np.arcsin(X)
plt.plot(X, Finv[0])
plt.plot(X, Finv[1])

cos2 = 1 - X * X

kn1 = np.array(X)
kn = (2 / np.pi) * (X * np.sqrt(cos2))

for n in range(2, MAXN):
    Finv[n] = kn + Finv[n-2]
    kn1, kn = kn, ((n-1) / n) * (cos2 * kn1)
    
    plt.plot(X, Finv[n])
plt.show()
```

```python
from math import pi, tau
N = 51

chosen = [[] for _ in range(N)]
chosen[1] = [1]
V = [1.0] * N
V[1] = 2.0
T = [0.0] * N
T[1] = 1.0
power2 = 2
for n in range(2, N):
    power2 *= 2
    V[n] = (tau * V[n-2]) / n
    T[n] = float("inf")
for n in range(2, N):
    it = [0] * n
    it[-1] = n
    sumit = [0] * n
    sumit[-1] = n
    maxit = [1] * n
    for i in range(n):
        maxit[i] = n // (n-i)
    it, sumit, maxit
    while True:
        i = n - 2
        while it[i] == maxit[i]:
            i -= 1
        it[i] += 1
        sumit[i] += 1
        for j in range(i+1, n-1):
            it[j] = it[j-1]
            sumit[j] = sumit[j-1] + it[j]
            maxit[j] = (n - sumit[j-1]) // (n-j)
        it[-1] = n - sumit[-2]

        # stuff
        t = 0.0
        v = 1 / V[n]
        start = 0
        while not it[start]:
            start += 1
        for k in range(start, n):
            v *= V[it[k]]
            t += T[it[k]]
        t *= v
        if T[n] > t:
            T[n] = t
            chosen[n] = list(it[start:])
        
        if it[0]:
            break
for i in range(N):
    print(i, chosen[i])
```

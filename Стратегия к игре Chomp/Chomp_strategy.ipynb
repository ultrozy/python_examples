```python
import numpy as np
from copy import deepcopy
OnlyRect = True
N = 3
M = 30
```

```python
def nextind(ind: list):
    pos = 0
    while (pos < N) and (ind[pos] == M-1):
        ind[pos] = 0
        pos += 1
    if pos < N:
        ind[pos] += 1
    if OnlyRect:
        while sum(ind) >= M:
            pos = 0
            while (pos < N) and (ind[pos] == M-1):
                ind[pos] = 0
                pos += 1
            if pos < N:
                ind[pos] += 1
    return ind
```

```python
shap = []
for i in range(N):
    shap = shap + [M]
losepos = np.full(tuple(shap), False)
curind = [1]
for i in range(N-1):
    curind = curind + [0]
```

```python
while sum(curind) > 0:
    
    if sum(curind) == curind[N-1]:
        print("Done for {}".format(curind))
    
    found = False
    
    ind = deepcopy(curind)
    pos = 0
    while True:
        while (pos < N) and (ind[pos] == 0):
            pos += 1
        if pos < N:
            ind[pos] -= 1
        else:
            break
        if losepos[tuple(ind)]:
            found = True
            break
    if found:
        curind = nextind(curind)
        continue
    
    for start in range(N-1):
        ind = deepcopy(curind)
        pos = start + 1
        while True:
            while (pos < N) and (ind[pos] == 0):
                pos += 1
            if pos < N:
                ind[pos] -= 1
                ind[start] += 1
            else:
                break
            if losepos[tuple(ind)]:
                found = True
                break
        if found:
            break
    
    if not found:
        losepos[tuple(curind)] = True
    curind = nextind(curind)
```

```python
curind = nextind([0] * N)
while sum(curind) > 0:
    if losepos[tuple(curind)]:
        print(list(np.flip(np.cumsum(np.flip(np.array(curind))))), "\n")
        #print(curind, "\n")
    curind = nextind(curind)
```
